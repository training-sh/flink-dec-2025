```python
from pyflink.datastream import StreamExecutionEnvironment, RuntimeExecutionMode
from pyflink.common.typeinfo import Types
from pyflink.table import StreamTableEnvironment
```

```python

# Tasks run on Remote manager, uses job manager, task manager, environment, task manager, job manager runs runs in the cluster.
# by default, there will be one ACTIVE job manger, as many task managers
# get_env is just a helper menthod
# flink client (this notebook), job manager , task manager runs on differet machines
import get_env
env = get_env.get_remote_env()
t_env = StreamTableEnvironment.create(env)
t_env.get_config().set("parallelism.default", "1")
```

```python
data = list(range(10))
# Tiny pipeline
# Flink is structured, need data type, schema
ds = env.from_collection(data, type_info=Types.INT())
# LAZY EVAL
even = (ds
        .map(lambda x: x + 1, output_type=Types.INT())
        .filter(lambda x: x % 2 == 0))
```

```python
# Printing to stdout just to have a running sink:
even.print() # does not print, it create a simple DAG, discussed in next notebook
```
